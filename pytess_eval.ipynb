# Pytesseract Evaluation

## Import some libraries...

```python
# import general libraries
import pandas as pd
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
import cv2
import string
import time
```

```python
# import libraries for connecting to S3
import boto3 
import botocore 
from sagemaker import get_execution_role 
```

```python
# import libraries for pytesseract 
import pytesseract
from PIL import Image
from pytesseract import Output
```

## Connect to S3 and read 6k eval data...

```python
# connect to data in S3 
role = get_execution_role() 
bucket = 'sagemaker-060720' 
data_key = 'evalOFFdata.csv' 
data_location = 's3://{}/{}'.format(bucket, data_key) 
```

```python
# load 6k eval data
data = pd.read_csv(data_location)

data.shape
```

```python
start_time = time.time()
```

## Preprocess for OCR...

```python
# np arrays with RGB
imgs = []

for image_url in data.image_ingredients_url[0:10]:
    img = io.imread(image_url)
    imgs.append(img)
```

## Run OCR and show images...

```python
# words in the box
texts = []

# position on the page (in pixels)
boxes = []

for i in range(10):
    img = imgs[i].copy()
    print("----------"+str(i)+"----------")
    print("_____ Original _____")
    plt.imshow(img)
    plt.show()
    box = pytesseract.image_to_data(img, output_type=Output.DICT)
    text = box['text']
    print("_______ Text _______")
    print(text)
    texts.append(text)
    n_boxes = len(box['level'])
    for i in range(n_boxes):
        (x, y, w, h) = (box['left'][i], box['top'][i], box['width'][i], box['height'][i])
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2) 
    print("____ Recognized ____")
    plt.imshow(img)
    plt.show()
    boxes.append(box)
    
print("--- %s seconds ---" % (time.time() - start_time))
```

## Evaluate OCR results...

```python
ocr_ingredients = []

for i in range(10):
    ocr_ingredient = [s.translate(str.maketrans('', '', string.punctuation)).lower() for s in texts[i]]
    ocr_ingredients.append(ocr_ingredient)
```

```python
ingredients = []

for i in range(10):
    ingredient = [s.translate(str.maketrans('', '', string.punctuation)).lower() for s in data.ingredients_text[i].split()]
    ingredients.append(ingredient)
```

```python
# number of actual ingredients in predictions
for i in range(10):
    print(sum([ing in ocr_ingredients[i] for ing in ingredients[i]]))
```
